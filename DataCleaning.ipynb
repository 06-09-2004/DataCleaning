```python
import pandas as pd
from pandas import read_csv
path="/content/drive/MyDrive/rawdata.csv"
df=pd.read_csv(path)
new_df=df.dropna()
print(new_df)

```

```python
new_df=df.dropna()
print(new_df.to_string())
```

```python
#replace null values
new_df=df.fillna(130)#null values
print(new_df.to_string())
```

```python
#replace only for specified columns
new_df=df["Calories"].fillna(130);
print(new_df.to_string())
```

```python
#replace only for specified columns
new_df=df["Duration"].fillna(130);
print(new_df.to_string())
```

```python
#replace using mean, median, mode
x=df["Duration"].mean()
new_df=df["Duration"].fillna(x);
print(new_df.to_string)
#df["Duration"].fillna(x, implace=True)

y=df["Pulse"].median
new_df=df["Pulse"].fillna(y);
print(new_df.to_string)
#df["pulse"].fillna(y, inplace= True)

z=df["Maxpulse"].median
new_df=df["Maxpulse"].fillna(z);
print(new_df.to_string)
#df["Maxpulse"].fillna(z, inplace= True)



```

```python
#fixing wrong rawdata
df.loc[7,'Duration']=45
print(df.to_string())
#Disadvantage is we have to know row, number
```

```python
#Loop through all values in the "Duration" column
for  x in df.index: #membership function(in)
  if df.loc[x, "Duration"]>120:
    df.loc[x, "Duration"]=120
print(df.to_string())

```
